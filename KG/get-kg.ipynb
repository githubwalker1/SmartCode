```python
'''
将kg_num中的三元组处理为字典列表
'''
# 定义一个函数来处理文件
def process_triples(file_path):
    # 创建一个字典来存储结果
    result = {}

    # 打开并读取文件
    with open(file_path, 'r',encoding='utf-8') as file:
        for line in file:
            # 假设每行是由逗号分隔的三个元素
            line = line.strip().strip('()')
            elements = line.split(',')

            # 检查是否有三个元素
            if len(elements) != 3:
                continue

            # 解包三元组
            first, second, third = elements
            second = second.lstrip()
            third = third.lstrip()
            # 更新字典
            if first not in result:
                result[first] = {third: [int(second)]}
            else:
                if third in result[first]:
                    result[first][third].append(int(second))
                else:
                    result[first][third] = [int(second)]

    # 将字典转换为所需的列表格式
    result_list = [{k: v} for k, v in result.items()]

    return result_list

# 假设您的文件名为'triples.txt'
file_path = 'kg_num'
result = process_triples(file_path)
for i in result:
    print(i)
```

```python
'''
将三元组列表result写入merged_dict.json文件中。
'''
import json

# 合并字典
merged_dict = {}
for d in result:
    for key, value in d.items():
        if key in merged_dict:
            # 如果键已存在，则更新其值
            merged_dict[key].update(value)
        else:
            # 如果键不存在，添加新的键值对
            merged_dict[key] = value

# 写入JSON文件
output_file = 'merged_dict.json'
with open(output_file, 'w') as file:
    json.dump(merged_dict, file, indent=4)

# 输出文件路径
print(f"Merged dictionary written to {output_file}")
```
